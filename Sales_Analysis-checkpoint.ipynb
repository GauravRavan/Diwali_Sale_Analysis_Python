# State wise observations

```python
df.columns
```

```python
StatebyOredrCount = df.groupby(["State"],as_index= False)["Orders"].sum().sort_values(by="Orders", ascending = False).head(10) #this head(10) gives only top 10 by defualt only head() gives 5
# StatebyOredrCount
s.set(rc ={"figure.figsize":(18,5)})
x = s.barplot(x="State", y = "Orders", data = StatebyOredrCount)
for bars in x.containers:
    x.bar_label(bars)
```

```python
StateByAmountSpend = df.groupby(["State"],as_index=False)["Amount"].sum().sort_values(by="Amount", ascending = False).head(10)
s.set(rc ={"figure.figsize":(20,5)})

sbas = s.barplot(x="State", y="Amount", data  =StateByAmountSpend)
```

### Observations from the Bar Plot Analysis

1. **State Order and Amount Spent Distribution**  
   - The majority of orders and the highest spending come from **Uttar Pradesh, Maharashtra, and Karnataka**.

2. **State-Specific Spending Patterns**  
   - **Haryana** has a smaller number of orders but a higher total spending, while **Kerala** is in the top 10 for the number of orders, yet the overall amount spent is comparatively lower.

### Overall Conclusion from Diwali Sale Analysis

- **Key Demographic**: During this Diwali sale, **females**, particularly **married females** in the **26-35 age group**, were the highest spenders.
- **Top Spending Regions**: The majority of spending came from customers in **Uttar Pradesh**, **Maharashtra**, and **Karnataka**.
- **Top Professions**: Individuals working in **IT**, **banking**, and **aviation** sectors made the most purchases and spent the most money.
- **Preferred Categories**: The highest expenditures were on **food**, **clothing**, and **electronics**, showing these as the most popular categories.
- **Regional Insights**: The **Central** zone led in both the number of orders and total amount spent, making up a significant **39%** of all sales.
